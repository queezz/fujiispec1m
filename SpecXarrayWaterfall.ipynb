```python
import os
from os.path import join, exists
import xarray as xr
```

```python
basepath = r'\\10.249.254.51\Public\Kuzmin\Experiments-Leprecon\Spectrometer\20200821'
ls = os.listdir(basepath)
```

```python
rcParams['font.size'] = 20
```

```python
ls = os.listdir(basepath)
ls = [i for i in ls if 'N2' in i]
step = 5e6
fig=gcf()
fig.set_facecolor('w')
fig.set_size_inches([30,45])

for i,f in enumerate(ls[:]):
    fp = join(basepath, f)
    data = xr.open_dataset(fp)
    image_data = data.to_array()[0]
    plot(image_data.sum(axis=0)+i*step,'k')


[text(-120,2.6e6+step*i,t[3:-4],color='r') for i,t in enumerate(ls)]
xlim(-170,2100)

xlabel('pixel')
ylabel('intensity')
#savefig('../img/hespectra.pdf',dpi=200,bbox_inches='tight')
```
